```
creditcard = data.iloc[:,:30]
target= data['Class']

print('Shape of X is', creditcard.shape)
print('Shape of y is', target.shape)
```

```
creditcard.head()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(creditcard, target, test_size = 0.2, random_state = 10)
```

```
print('Shape of X_train = ', X_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of y_test = ', y_test.shape)

c = y_test
```

```
c
```

## Feature Selection

```
from sklearn.feature_selection import VarianceThreshold
```

```
var_threshold = VarianceThreshold(threshold = 0)

#var_threshold.fit(X_train)
```

## Feature Scaling

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

## Model Evalution

```
# impoer requiede packages
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

## Decision Tree

```
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(criterion= 'entropy', random_state=0)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)

a_dt = accuracy_score(y_test, y_pred_dt)
a_dt
```

```
# train with Standert Scaling dataset
dt_model2 = DecisionTreeClassifier(criterion= 'entropy', random_state=0)
dt_model2.fit(X_train_sc, y_train)
y_pred_dt_sc = dt_model2.predict(X_test_sc)

a_dt_sc = accuracy_score(y_test, y_pred_dt_sc)

a_dt_sc
```

KNN

```
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2,)
knn_model.fit(X_train, y_train)
y_pred_knn = knn_model.predict(X_test)

a_knn = accuracy_score(y_test, y_pred_knn)
a_knn
```

```
# train with Standert Scaling dataset
knn_model2 = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2,)
knn_model2.fit(X_train_sc, y_train)
y_pred_knn_sc = knn_model2.predict(X_test_sc)

a_knn_sc = accuracy_score(y_test, y_pred_knn_sc)
a_knn_sc
```

### Naive Bayes

```
# Naive Bayes
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)

a_n = accuracy_score(y_test, y_pred_nb)
a_n
```

```
# train with Standert Scaling dataset
nb_model2 = GaussianNB()
nb_model2.fit(X_train_sc, y_train)
y_pred_nb_sc = nb_model2.predict(X_test_sc)

a_n_sc =  accuracy_score(y_test, y_pred_nb_sc)
a_n_sc
```

Random Forest Classifier

```
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

a_rd =accuracy_score(y_test, y_pred_rf)

a_rd
```

```
# train with Standert Scaling dataset
rf_model2 = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
rf_model2.fit(X_train_sc, y_train)
y_pred_rf_sc = rf_model2.predict(X_test_sc)

a_rd_sc = accuracy_score(y_test, y_pred_rf_sc)

a_rd_sc
```

### Support Vector Machine

```
# Support Vector Machine
from sklearn.svm import SVC
svc_model = SVC()
svc_model.fit(X_train, y_train)
y_pred_svc = svc_model.predict(X_test)

a_svc = accuracy_score(y_test, y_pred_svc)
a_svc
```

```
# train with Standert Scaling dataset
svc_model2 = SVC()
svc_model2.fit(X_train_sc, y_train)
y_pred_svc_sc = svc_model2.predict(X_test_sc)

a_svc_sc = accuracy_score(y_test, y_pred_svc_sc)
a_svc_sc
```

### XGBoost Classifier

```
from xgboost import XGBClassifier
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)

a_xgb = accuracy_score(y_test, y_pred_xgb)
a_xgb
```

```
# train with Standert Scaling dataset
xgb_model2 = XGBClassifier()
xgb_model2.fit(X_train_sc, y_train)
y_pred_xgb_sc = xgb_model2.predict(X_test_sc)

a_xgb_sc = accuracy_score(y_test, y_pred_xgb_sc)
a_xgb_sc
```
